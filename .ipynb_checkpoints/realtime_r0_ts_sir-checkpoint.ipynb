## Visualizando a evolução do número de reprodução no Brasil (séries temporais e SIR) 

Vamos tentar modelar o número de reprodução eficaz usando um modelo de séries temporais ajustado no crescimento percentual diário e a premissa do SIR de que 

$$\large R_t = \frac{\beta}{\gamma}$$

onde $\beta$ é a taxa de taxa de crescimento diária (que vamos modelar) e $\gamma$ é a taxa de recuperação ou a recíproca do tempo até a recuperação.

Começamos importando as bibliotecas necessárias. Condensamos boa parte do código no script `core.py`, para facilitar a utilização desse notebook.

```python
# importing core 
# aqui empacotamos as funções que dá pra achar no código do Kevin Systrom
from core import run_full_model, fit_gam, run_gam_effective_r, load_data, plot_rt, plot_standings

# imports básicos
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

# importando joblib para rodar em paralelo
from joblib import Parallel, delayed

from pygam import GammaGAM, PoissonGAM, InvGaussGAM, s, l

# para plots com uma melhor resolução
%config InlineBackend.figure_format = 'retina'
```

Área reservada para parâmetros do modelo. Até agora só deixamos aberto `N_JOBS`, para escolher o número de núvleos para o ajuste dos modelos.

```python
### parâmetros ###

# númreo de cores para paralelizar modelo nos estados
N_JOBS = -1
```

### Leitura dos dados

Usamos o [repositório do Wesley Cota](https://github.com/wcota/covid19br) para pegar os dados mais atualizados por estado. 

```python
### reading data ###

# lendo do repositório do Wesley Cota
city_df, state_df = load_data()

# para a análise, vamos usar casos confirmados acumulados
#state_df = state_df['confirmed_total'].groupby(level='state').pct_change().dropna()
#state_df = state_df['confirmed_total']
```

### Ajuste do modelo

Usamos a biblioteca `pygam`, que implementa Modelos Generalizados Aditivos (GAMs) que podem ser bastante expressivos, usando B-splines penalizadas como *feature functions*, o que nos permite capturar relações não lineares. No nosso caso, como estamos modelando séries temporais, podemos enxergar a expressividade do modelo como memória: quanto mais expressivo o modelo, menos memória ele terá, e usará informação mais recente.

Outro benefício de GAMs é permitir a modelagem de distribuições de erro não gaussianas. No nosso caso, isto é importante: como estamos modelando crescimento percentual de uma variável que é acumulada, uma distribuição normal para o erro não deve funcionar já que não podemos admitir valores menores que 0. Em geral, a literatura (inclusive a epidemiológica) sugere o uso de uma distribuição *Gamma*.

```python
### running posteriors ###

with Parallel(n_jobs=N_JOBS) as parallel:
    results = parallel(delayed(run_gam_effective_r)(grp[1], 20) for grp in state_df.groupby(level='state'))

final_results = pd.concat(results)
```

### Observando os resultados: $R_t$

Usamos o mesmo padrão de visualização adotado pelo Kevin. Nos gráficos, cada ponto é a estimativa de $R_t$ para o dia, e a área cinzenta indica o intervalo crível de 90% de probabilidade. Deixamos $R = 1$ em evidência, já que quando $R_t < 1$ indica que a epidemia está sob controle.

```python
### plotting - Rt vs time for all states ###

def plot_rt_states():

    # number of columns and rows for plotting
    N_COLS = 4
    N_ROWS = int(np.ceil(len(results) / N_COLS))

    # opening figure
    fig, axes = plt.subplots(nrows=N_ROWS, ncols=N_COLS, figsize=(15, N_ROWS*3), dpi=90)

    # loop for several states
    for i, (state_name, result) in enumerate(final_results.groupby('state')):
        plot_rt(result, axes.flat[i], state_name)

    # saving figure
    fig.tight_layout()
    fig.set_facecolor('w')

plot_rt_states()
```

### Observando os resultados: comparação entre estados

Para fazer uma comparação entre estados, mostramos a última estimativa de $R_t$ de cada estado no gráfico de barras a seguir, com a incerteza associada. Os gráficos estão ordenados do melhor para o pior usando a estimativa mais provável do modelo.

```python
### plotting - state comparison ###

mr = final_results.groupby(level=0)[['ML', 'High_90', 'Low_90']].last()
mr.sort_values('ML', inplace=True)
plot_standings(mr, figsize=(13,5));
```

Podemos ordenar também pelo pior caso possível também, usando nossas estimativas de incerteza:

```python
### ordering by worst case ##

mr.sort_values('High_90', inplace=True)
plot_standings(mr, figsize=(13,5));
```
